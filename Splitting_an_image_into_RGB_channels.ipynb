```
path="/content/Baby.jpg"
from google.colab.patches import cv2_imshow
```

```
import cv2
import numpy as np
```

```
baby=cv2.imread(path)
(width,height,channels)=np.shape(baby)
print(width,height,channels)
```

```
cv2_imshow(baby[:,:,2]) # red channel
```

As the red channel slice of image is a 2D matrix and 2D ranges from (0 - 255). That is the reason we see them in grayscale. Even thought the pixels are of red channels, we can only see them in grayscale

Now lets actually print the red, green and blue channels (not the grayscale version)

```
r = baby.copy()
r[:, :, 0] = 0
r[:, :, 1] = 0
g = baby.copy()
g[:, :, 0] = 0
g[:, :, 2] = 0
b = baby.copy()
b[:, :, 1] = 0
b[:, :, 2] = 0
horizontal= np.hstack((baby,r,g,b))
cv2_imshow(horizontal)
```
