```
img1="/content/shapes.jpeg"
img2="/content/car.jpeg"
```

```
import numpy as np
import matplotlib.pyplot as plt
# defining filters

image=plt.imread(img1)
fig,ax=plt.subplots()
ax.imshow(image)
ax.axis('off')
plt.show()
```

```
h,w,c=image.shape # dimesnions of the image
#make a copy of image 
h_image1=image.copy()
h_image2=image.copy()
h_image3=image.copy()
h_image4=image.copy()
h_image=image.copy()
```

```
vertical_filter=np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
# as filter size is 3x3

for i in range(0,h-2):
  for j in range(0,w-2):
    local=image[i:i+3,j:j+3,0]
    #print("local",local)
    hor_conv=np.multiply(vertical_filter,local)
    #print("hor_conv",hor_conv)
    h_image1[i,j]=np.sum(hor_conv)/128
    #print("h_image",h_image[i,j])
fig,ax=plt.subplots()
ax.imshow(h_image1)
ax.axis('off')
plt.show()
```

```
vertical_filter=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
# as filter size is 3x3

for i in range(0,h-2):
  for j in range(0,w-2):
    local=image[i:i+3,j:j+3,0]
    #print("local",local)
    hor_conv=np.multiply(vertical_filter,local)
    #print("hor_conv",hor_conv)
    h_image2[i,j]=np.sum(hor_conv)/128
    #print("h_image",h_image[i,j])
fig,ax=plt.subplots()
ax.imshow(h_image2)
ax.axis('off')
plt.show()
```

```
horizontal_filter=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
# as filter size is 3x3

for i in range(0,h-2):
  for j in range(0,w-2):
    local=image[i:i+3,j:j+3,0]
    #print("local",local)
    hor_conv=np.multiply(horizontal_filter,local)
    #print("hor_conv",hor_conv)
    h_image3[i,j]=np.sum(hor_conv)/128
    #print("h_image",h_image[i,j])
fig,ax=plt.subplots()
ax.imshow(h_image3)
ax.axis('off')
plt.show()
```

```
horizontal_filter=np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
# as filter size is 3x3

for i in range(0,h-2):
  for j in range(0,w-2):
    local=image[i:i+3,j:j+3,0]
    #print("local",local)
    hor_conv=np.multiply(horizontal_filter,local)
    #print("hor_conv",hor_conv)
    h_image4[i,j]=np.sum(hor_conv)/128
    #print("h_image",h_image[i,j])
fig,ax=plt.subplots()
ax.imshow(h_image4)
ax.axis('off')
plt.show()
```

```
# using SOBEL FILTER ref:https://en.wikipedia.org/wiki/Sobel_operator#:~:text=The%20Sobel%20operator%2C%20sometimes%20called,creates%20an%20image%20emphasising%20edges.
# horizontal 
vertical_filter=np.array([[-1,0,-1],[-2,0,2],[-1,0,1]])
horizontal_filter=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
# as filter size is 3x3

for i in range(0,h-2):
  for j in range(0,w-2):
    local=image[i:i+3,j:j+3,0]
    #print("local",local)
    hor_conv=np.multiply(horizontal_filter,local)
    ver_conv=np.multiply(vertical_filter,local)
    #print("hor_conv",hor_conv)
    hor_score=np.sum(hor_conv)/4
    ver_score=np.sum(ver_conv)/4
    h_image[i,j]=(hor_score**2+ver_score**2)**(0.5)
h_image=h_image/h_image.max()
fig,ax=plt.subplots()
ax.imshow(h_image)
ax.axis('off')
plt.show()
```
